```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets, metrics
from sklearn.preprocessing import StandardScaler

# main statement
if __name__ == '__main__':
    # loading dataset
    x, y = datasets.load_boston(return_X_y=True)
    # separating training and testing data
    x_temp = np.zeros([x.shape[0], x.shape[1] + 1])
    x_temp[:, 0] = 1
    x_temp[:, 1:] = x
    x_train = x_temp[0:400, :]
    y_train = y[0:400]
    x_test = x_temp[400:506, :]
    y_test = y[400:506]
    # instance of StandardScalar
    sc = StandardScaler()
    x_train[:, 1:] = sc.fit_transform(x_train[:, 1:])
    x_test[:, 1:] = sc.fit_transform(x_test[:, 1:])
    L = 0.001  # alpha
    n = x_train.shape[0]
    epochs = 10000  # epochs
    Frame = pd.DataFrame(
        columns=["Lambda", "theta0", "theta1", "theta2", "theta3", "theta4", "theta5", "theta6", "theta7", "theta8",
                 "theta9", "theta10", "theta12", "theta13", "MAE", "MSE"])
    Lambda = [5, 10, 15, 20, 25, 30]  # different lambdas
    k = 0
    theta_j = [[], [], [], [], [], []]
    for i in Lambda:
        theta = np.zeros((x_train.shape[1]))
        for _ in range(epochs):
            y_predictions = np.dot(x_train, theta)
            error = y_predictions - y_train
            theta[0] = theta[0] - (L * ((1 / n) * np.dot(error, x_train[:, 0])))
            for j in range(1, x_train.shape[1]):
                theta[j] = (theta[j] * (1 - ((L * i) / n))) - (L * ((1 / n) * np.dot(error, x_train[:, j])))
            cost_function = (1 / (2 * n) * (np.dot(error.T, error) + i * np.dot(theta.T, theta)))
            theta_j[k].append(cost_function)
        k = k + 1
        y_prediction = np.dot(x_test, theta)
        MAE = metrics.mean_absolute_error(y_true=y_test, y_pred=y_prediction)
        MSE = metrics.mean_squared_error(y_true=y_test, y_pred=y_prediction)
        Frame = Frame.append({'Lambda': i, 'theta 0': theta[0], 'theta 1': theta[1], 'theta 2': theta[2],
                              'theta 3': theta[3], 'theta 4': theta[4], 'theta 5': theta[5], 'theta 6': theta[6],
                              'theta 7': theta[7], 'theta 8': theta[8], 'theta 9': theta[9], 'theta 10': theta[10],
                              'theta 11': theta[11], 'theta 12': theta[12], 'theta 13': theta[13],
                              'MAE': MAE, 'MSE': MSE},
                             ignore_index=True)
    print(Frame)
    plt.plot(range(epochs), theta_j[0], color="red", label="Lambda 5")
    plt.plot(range(epochs), theta_j[1], color="green", label="Lambda 10")
    plt.plot(range(epochs), theta_j[2], color="blue", label="Lambda 15")
    plt.plot(range(epochs), theta_j[3], color="yellow", label="Lambda 20")
    plt.plot(range(epochs), theta_j[4], color="black", label="Lambda 25")
    plt.plot(range(epochs), theta_j[5], color="orange", label="Lambda 30")
    plt.xlabel("Iterations")
    plt.ylabel("Cost Function")
    plt.legend()
    plt.show()
```

```python
# importing libraries
import pandas as pd
import numpy as np
from sklearn import datasets, metrics
from numpy.linalg import inv
import matplotlib.pyplot as plt

# main statement
if __name__ == '__main__':
    # loading dataset
    x, y = datasets.load_boston(return_X_y=True)
    # separating training and testing data
    x_temp = np.zeros([x.shape[0], x.shape[1] + 1])
    x_temp[:, 0] = 1
    x_temp[:, 1:] = x
    x_train = x_temp[0:400, :]
    y_train = y[0:400]
    x_test = x_temp[400:506, :]
    y_test = y[400:506]
    theta = np.zeros(x_train.shape[1])
    Frame = pd.DataFrame(
        columns=["theta0", "theta1", "theta2", "theta3", "theta4", "theta5", "theta6", "theta7", "theta8", "theta9",
                 "theta10", "theta12", "theta13", "MAE", "MSE"])
    alpha = 10
    # normalization process started
    Lambda = np.zeros((x_train.shape[1], x_train.shape[1]))
    for i in range(1, x_train.shape[1]):
        Lambda[i, i] = alpha
    XTX1 = inv((np.dot(x_train.T, x_train)) + Lambda)
    XTy = np.dot(x_train.T, y_train)
    theta = np.dot(XTX1, XTy)
    y_prediction = np.dot(x_test, theta)
    MAE = metrics.mean_absolute_error(y_true=y_test, y_pred=y_prediction)
    MSE = metrics.mean_squared_error(y_true=y_test, y_pred=y_prediction)
    Frame = Frame.append({'theta0': theta[0], 'theta1': theta[1], 'theta2': theta[2], 'theta3': theta[3],
                          'theta4': theta[4], 'theta5': theta[5], 'theta6': theta[6], 'theta7': theta[7],
                          'theta8': theta[8], 'theta9': theta[9], 'theta10': theta[10], 'theta11': theta[11],
                          'theta12': theta[12],
                          'theta13': theta[13], 'MAE': MAE, 'MSE': MSE},
                         ignore_index=True)
    print(Frame)
    plt.scatter(range(x_test.shape[0]), y_prediction)
    plt.scatter(range(x_test.shape[0]), y_test, color='red')
    plt.show()
```
